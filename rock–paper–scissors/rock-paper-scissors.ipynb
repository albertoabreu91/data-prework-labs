```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list
import random

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.

outcomes = ['stone', 'paper', 'scissors']

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...

max_games = random.choice(list(range(1,10,2)))


# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games

games_to_win = (max_games // 2) + 1

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.

mac_choice = ''

def machine_choice():
    global mac_choice
    mac_choice = random.choice(outcomes)
    return mac_choice

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.

your_choice = ''

def u_choice():
    global your_choice
    while True:
        try:
            your_choice = str.lower(input('What is your choice: '))

        except ValueError:
            print("Sorry, I didn't understand that.")
            continue

        if your_choice not in outcomes:
            print("Sorry, your response must 'stone','paper' or 'scissors' \n")
            continue
        else:
            break
  
    return your_choice


# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins

dict_combats = {
  'paper' : 'stone',
  'stone' : 'scissors',
  'scissors' : 'paper',
}


outcome_per_combat = 0

def combat():
    machine_choice()
    u_choice()
    global mac_choice
    global your_choice
    global outcome_per_combat
    if dict_combats[your_choice] == mac_choice:
        outcome_per_combat = 2
    if dict_combats[mac_choice] == your_choice:
        outcome_per_combat = 1
    if dict_combats[mac_choice] == dict_combats[your_choice]:
        outcome_per_combat = 0

    return outcome_per_combat

    
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated

def state_game():
    global mac_wins
    global user_wins

    combat()
    print('Machine choice:', mac_choice)
  
    if outcome_per_combat == 1:
        print('Machine wins this round')
        mac_wins += 1
    
    if outcome_per_combat == 2:
        print('User wins this round')
        user_wins +=1
    
    elif outcome_per_combat == 0:
        print('Its a tie')
    
    print('Total games:', max_games)
    print('Minimun games to win:', games_to_win)
    print('Machine:', mac_wins, 'User:', user_wins)
    print('\n')

# Create two variables that accumulate the wins of each participant

mac_wins = 0
user_wins = 0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.

def main():
    while (mac_wins < games_to_win) and (user_wins < games_to_win):
        if (mac_wins < games_to_win) or (user_wins < games_to_win):
            state_game()
        else:
            break
  
    return mac_wins, user_wins
  

main()
    
# Print by console the winner of the game based on who has more accumulated wins
  
if mac_wins > user_wins:
    print('The Machine wins the game!!!!!')

else: 
    print('The User wins the game!!!!!')
```
